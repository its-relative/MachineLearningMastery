Now that we have all the continuous columns and the categorical columns in place lets split the dataset.

### Splitting the dataset into test and train subsets

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.2,random_state=42)
```

### Lets Scale the data

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
# for i in con_cols:
x_train[con_cols] = sc.fit_transform(x_train[con_cols])
x_test[con_cols] = sc.transform(x_test[con_cols])
```

### Cooks distance and influence plots

```python
import statsmodels.api as sm
import matplotlib.pyplot as plt
from statsmodels.stats.outliers_influence import OLSInfluence

# Adding a constant term for intercept
x_train = sm.add_constant(x_train)

# Fit the linear regression model on the training set
model = sm.OLS(y_train, x_train).fit()

# Calculate Cook's distance and other influence statistics for training set
influence = OLSInfluence(model)
cook_distance_train = influence.cooks_distance[0]  # Extract the first element from the tuple

# Plotting influence plots for the training set
fig, ax = plt.subplots(figsize=(12, 8))

# Influence plot
sm.graphics.influence_plot(model, ax=ax, criterion="cooks")

plt.figure(figsize=(10, 6))
plt.stem(cook_distance_train, markerfmt=",", linefmt="-", basefmt=" ", label="Cook's Distance")
plt.axhline(4 / len(y_train), color="r", linestyle="--", label="Threshold")
plt.xlabel("Observation (Training Set)")
plt.ylabel("Cook's Distance")
plt.legend()
plt.title("Cook's Distance Plot for Training Set")
plt.show()
```

### Lets train the model on the data

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
```

Let's predict the test data and compare it with actual data

```python
y_pred = lr.predict(x_test)
check_df = pd.DataFrame({"pred":y_pred,"actual":y_test})
```
