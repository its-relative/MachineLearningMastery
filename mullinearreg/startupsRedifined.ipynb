### Let us split the data into x and y for further splitting into test and train

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1:]
```

```python
x
```

```python
y
```

### Let us split the dataset into test and train subsets

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

Let us OneHotEncode the categorical column i.e. State

```python
from sklearn.preprocessing import OneHotEncoder
oe = OneHotEncoder()
```

### Encoding

```python
x_train = pd.get_dummies(x_train).astype(int)
x_test = pd.get_dummies(x_test).astype(int)
```

```python
x_test
```

### Let us scale the data first before feeding it to the algorithm

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

Taking only the columns from x that don't have object as its datatype, because lets not forget that scaling is done only on continuous features.

```python
# con_cols = ["R&D Spend","Administration","Marketing Spend"]
con_cols = [col for col in x.columns if x[col].dtype != "object"]
con_cols
```

```python
x_train[con_cols] = sc.fit_transform(x_train[con_cols])
x_test[con_cols] = sc.transform(x_test[con_cols])
```

```python
x_train
```

```python
x_test
```

Now that we have scalled all the columns let us start regression

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

### Cook distance and Influence plot

```python
import statsmodels.api as sm
import matplotlib.pyplot as plt
from statsmodels.stats.outliers_influence import OLSInfluence

# Adding a constant term for intercept
x_train = sm.add_constant(x_train)

# Fit the linear regression model on the training set
model = sm.OLS(y_train, x_train).fit()

# Calculate Cook's distance and other influence statistics for training set
influence = OLSInfluence(model)
cook_distance_train = influence.cooks_distance[0]  # Extract the first element from the tuple

# Plotting influence plots for the training set
fig, ax = plt.subplots(figsize=(12, 8))

# Influence plot
sm.graphics.influence_plot(model, ax=ax, criterion="cooks")

plt.figure(figsize=(10, 6))
plt.stem(cook_distance_train, markerfmt=",", linefmt="-", basefmt=" ", label="Cook's Distance")
plt.axhline(4 / len(y_train), color="r", linestyle="--", label="Threshold")
plt.xlabel("Observation (Training Set)")
plt.ylabel("Cook's Distance")
plt.legend()
plt.title("Cook's Distance Plot for Training Set")
plt.show()
```
