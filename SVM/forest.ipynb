### Lets split the data

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1:]
```

```python
x
```

```python
y.value_counts()
```

Lets split the columns in continuous and categorical for better understanding

```python
con_cols = [col for col in x.columns if x[col].dtype != "object"]
cat_cols = [col for col in x.columns if x[col].dtype == "object"]
```

```python
y = y.replace("small",0)
y = y.replace("large",1)
```

```python
y
```

### Splitting in train and test

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

### Scaling and Encoding

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
x_train[con_cols] = sc.fit_transform(x_train[con_cols])
x_test[con_cols] = sc.transform(x_test[con_cols])
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.preprocessing import OneHotEncoder
oe = OneHotEncoder(handle_unknown="ignore")
```

```python
oe_train = pd.DataFrame(oe.fit_transform(x_train[cat_cols]).toarray(),index = x_train[cat_cols].index)
oe_test = pd.DataFrame(oe.transform(x_test[cat_cols]).toarray(),index = x_test[cat_cols].index)
```

```python
oe_train
```

```python
oe_test
```

Lets see how the test final and train final turns out

```python
train_final = pd.concat([x_train[con_cols],oe_train],axis =1)
test_final = pd.concat([x_test[con_cols],oe_test],axis = 1)
```

```python
train_final
```

```python
test_final
```

### Lets Train the Model

```python
from sklearn.svm import SVC
svc = SVC()
```

```python
train_final.columns = train_final.columns.astype(str)
test_final.columns = test_final.columns.astype(str)
```

```python
svc.fit(train_final,y_train)
```

```python
y_pred  = svc.predict(test_final)
```

Lets pack the actual and predicted values side by side for better understanding

```python
check_df = pd.DataFrame({"pred":y_pred,"actual":np.array(y_test).reshape(-1)})
```

```python
check_df
```

### Lets check the accuracy

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
